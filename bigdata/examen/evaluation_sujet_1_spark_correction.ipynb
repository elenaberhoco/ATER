## Bonus

Le supermarché Arvis souhaite effectuer des statistiques sur l'ensemble des achats de ses clients sur les douze derniers mois. L'entreprise possède plusieurs magasins. À chaque passage de client à la caisse, une ligne de log est sauvegardée dans un fichier. Cette ligne de log est formatée de la manière suivante :
* Le premier mot indique la date de l'achat au format `JJ_MM_AAAA`
* Le deuxième mot indique l'heure de l'achat au format `HH_MM_SS`
* Le troisième mot est l'identifiant de la transaction
* Le quatrième mot est l'identifiant du caissier qui a pris en charge le client
* Le reste de la ligne contient les achats, les produits étant séparés par le caractère `+`

Par exemple :  `01_11_2019 15_22_08 200004 C002 sucre+lait` indique que le caissier `C002` a encaissé un panier comportant les produits `sucre` et `lait` le 01/11/2019 à 15h22min08sec. La transaction a l'identifiant `200004`.

Écrire un code Spark permettant de **calculer par tranche horaire d'une demi-heure** (toutes journées confondues) le **nombre total de produits** achetés dans la tranche, **tous magasins confondus**. Appliquer le code aux fichiers `transactions_magasin_A.txt` et `transactions_magasin_B.txt`. Le résultat attendu est le suivant :
```
[('09h00 à 09h29', 27),
 ('09h30 à 09h59', 24),
 ('10h00 à 10h29', 27),
 ('10h30 à 10h59', 18),
 ('11h00 à 11h29', 35),
 ('11h30 à 11h59', 27),
 ('12h00 à 12h29', 25),
 ('12h30 à 12h59', 45),
 ('13h00 à 13h29', 47),
 ('13h30 à 13h59', 26),
 ('14h00 à 14h29', 38),
 ('14h30 à 14h59', 36),
 ('15h00 à 15h29', 36),
 ('15h30 à 15h59', 25),
 ('16h00 à 16h29', 20),
 ('16h30 à 16h59', 38)]
```

````python
# Déterminer la tranche horaire d'une demi-heure

def tranche_horaire(horaire):
  horaire = horaire.split('_')
  heure = horaire[0]
  min_start = (int(horaire[1]) // 30) * 30
  min_end = min_start + 29
  tranche = f"{heure}h{str(min_start).zfill(2)} à {heure}h{min_end}"
  return tranche

tranche_horaire('15_22_08')
````

````python
sc.wholeTextFiles('*.txt')\
  .flatMap(lambda x: x[1].split('\n'))\
  .map(lambda x: x.split(' '))\
  .map(lambda x: (tranche_horaire(x[1]),len(x[4].split('+'))))\
  .reduceByKey(lambda x,y : x + y)\
  .sortBy(lambda x: x[0])\
  .collect()
````
