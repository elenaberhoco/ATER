2. Complétez votre compréhension du jeu de données par quelques recherches supplémentaires si nécessaire.

**Gaz à effet de serre** : [Gouvernement du Québec](https://www.environnement.gouv.qc.ca/air/questce-ges.htm#:~:text=Ils%20sont%20appel%C3%A9s%20%C2%AB%20gaz%20%C3%A0,'ozone%20(O3).)
> Certains gaz présents naturellement dans l'atmosphère terrestre contribuent à retenir la chaleur près de la surface de la Terre. Ils sont appelés « gaz à effet de serre » (GES) et formés essentiellement de vapeur d'eau, de dioxyde de carbone (CO2 ou gaz carbonique), de méthane (CH4), de protoxyde d'azote (N2O) et d'ozone (O3).

**Unité** : **kTeq CO2**, "kilotonne équivalent CO2"   
Unité de mesure multi-gaz (méthane, CO2, …) des émissions de gaz à effet de serre. Cette unité permet de cumuler l'impact des différents gaz à effet de serre sur l'environnement, chaque gaz n'ayant pas le même pouvoir de réchauffement global. [Connaissance des Énergies](https://www.connaissancedesenergies.org/questions-et-reponses-energies/gaz-effet-de-serre-quest-ce-que-l-equivalent-co2)
> Le GIEC considère qu'une tonne de méthane (CH4) a un pouvoir de réchauffement global 28 fois plus élevé en moyenne qu'une tonne de CO2 sur une période de temps de 100 ans. Ainsi, chaque tonne de méthane est comptabilisée comme 28 tonnes d'équivalent CO2 dans les bilans des émissions de GES.
